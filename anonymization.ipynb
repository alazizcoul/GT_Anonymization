```python
# Lire le fichier Stata

data = pd.read_stata(file_path)
#Retrait des non-réponses totales
data = data[data.oui.isin(['Oui'])==True ]
data = data.replace('',None)


# df, meta = pyreadstat.read_dta(file_path,encoding='latin1')
dic = {
    'NH_A': 'Temps_Dormir_reposer',
    'NH_B': 'Temps_manger_boire',
    'NH_C': 'Temps_Soin__personnel',
    'NH_D': 'Temps_Couture_tissage',
    'NH_E': 'Temps_Faire_la_cuisine',
    'NH_F': 'Temps_Voyage',
    'NH_G': 'Temps_Regarder_la_TV_Ecouter_la_radio_Lire',
    'NH_H': 'Temps_Promenade',
    'NH_I': 'Temps_Activites_sociales_et_loisirs',
    'NH_J': 'Temps_Activités_religieuses',

    'NH_K': 'Temps_Ecole_devoirs',
    'NH_L': 'Temps_Travailler_comme_employé',
    'NH_M': 'Temps_Travailler_pour_sa_propre_affaire',
    'NH_O': 'Temps_Ferme_Pêche_Pisciculture',
    'NH_P': 'Temps_Travaux_domestiques',
    'NH_Q': 'Temps_Charges_des_enfants_adultes_et_vieilles_personnes',
    'NH_W': 'Temps_Autre_Travail',
    'NH_X': 'Temps_Autre_loisir'

        }

for lettre in "ABCDEFGHIJKLMOPQWX":
    data["NH_"+lettre] =0
    meta.column_names_to_labels["NH_"+lettre] = dic["NH_"+lettre]

data['NH'] = 0
data["bn450"]= 0
data["bn70"]= 0
data["cas3_0"] = 0
data["bn2345"] = 0

meta.column_names_to_labels['NH']= 'Temps_travail_total'
meta.column_names_to_labels['bn450']= 'Depression_ou_anxiete'
meta.column_names_to_labels['bn70']= 'Bien etre mental'
meta.column_names_to_labels['cas3_0']= 'Relation_avec_la_famille'
meta.column_names_to_labels['bn2345']= 'Relation_avec_les_autres'



travail_loisir = [ "h"+str(1+i) for i in range(24)]
New_loisir_variables = ['NH_A','NH_B','NH_C','NH_F','NH_G','NH_H','NH_I','NH_J','NH_X']
New_travail_variables = ['NH_E','NH_D','NH_K', 'NH_L','NH_M','NH_O','NH_P','NH_Q','NH_W', ]

a_completer_x = ["h16","bn27", "bn28","bn29","bn30","bn38","bn37"]
a_completer_x_x = ["mt3","mt4","mt10","mt11","mt14","mt17","mt18"]
modalites_x = ["q","e","c","i",'f','d','c']
modalites_x_x = ["d","i","e","h","f","c","e"]

depression = ['bn451', 'bn452', 'bn453', 'bn454', 'bn455', 'bn456',
              'bn457', 'bn458', 'bn459', 'bn460', 'bn461', 'bn462']

bien_etre_mental = ['bn71','bn72','bn73','bn74','bn75','bn76','bn77']

relation_avec_les_autres = ['bn2', 'bn3', 'bn4', 'bn5']
relation_avec_la_famille =  ['cas3_1', 'cas3_2', 'cas3_3',]

# Pré-calcul des lettres pour chaque modalité
lettres_dict = {mod: [chr(i) for i in range(ord('A'), ord(mod.upper()) + 1)] for mod in set(modalites_x + modalites_x_x)}

def verifier_lettre(lettre, ligne):
    return 'Oui' if ligne != '' and lettre in ligne else 'Non'


def score_aggregate(row,list_var,dic_mod,new_var):
  for var in list_var :
    if  row[var] in dic_mod:
      row [var] = dic_mod[row[var]]
    else :
      row [var] = None

  val_list = [row[var] for var in list_var ]
  if set(val_list) != {None}:
    row[new_var] = sum(i for i in val_list if i !=None )
  else:
    row[new_var] = None


def remplacer(row):

    #Variables à compléter
    for var, mod in zip(a_completer_x + a_completer_x_x, modalites_x + modalites_x_x):
        if pd.notna(row[var]):
            suffixes = lettres_dict[mod]
            if  var !=  'bn28':
                suffixes = suffixes + (["X_X"] if var in a_completer_x else ["X"])
            for lettre in suffixes:
                if var + ('_' if var in a_completer_x else "") + lettre.lower() in row:
                    row[var + ('_' if var in a_completer_x else "") + lettre.lower()] = verifier_lettre(lettre, row[var])
                #print([var + ('_' if var in a_completer_x else "")+ lettre.lower()])

    #Correction des NA pour les quanti
    for var in ["bn6","bn7"]:
        if row[var]==998:
            row[var] = None

    #Duree de residence
    row["cj5"]= 2024-row["cj5"]

    #Temps de travail et de loisir
    for var in travail_loisir:
        if row[var+'x'] in travail_activites:
            row[var] = row[var].replace('R','W')
        elif row[var+'x'] in loisirs_activites:
            row[var] = row[var].replace('R','X')
        else :
            pass

    #
    for lettre in "ABCDEFGHIJKLMOPQWX":
        abcd_ = []
        for i in range(1,25):
          if row['h'+f'{i}'] != None:
            abcd_.append( (lettre in row['h'+f'{i}'] )/len(row['h'+f'{i}']) )
          else:
            abcd_.append(None)


        if set(abcd_) != {None}:
          row["NH_"+lettre] = sum(i for i in abcd_ if i is not None )
        else:
          row["NH_"+lettre] = None



    val_list = [row[var] for var in dic ]
    if set(New_loisir_variables+New_travail_variables) != {None}:
      row['NH'] = sum(row[var] for var  in New_loisir_variables if row[var] !=None)
    else:
      row['NH'] = None

    #Dépression ou anxiété
    score_aggregate( row=row,
                     list_var = depression,
                     dic_mod = {'    NON'               :0,
                                '   NON'                :0,
                                '\xa0\xa0 OUI'          :1,
                                '\xa0\xa0\xa0 OUI'      :1,
                                '\xa0\xa0\xa0\xa0 OUI'  :1
                                },
                     new_var = 'bn450'
                     )

    #Bien être mental
    score_aggregate( row=row,
                     list_var = bien_etre_mental,
                     dic_mod = {
                                'Jamais'              :1,
                                ' Rarement'           :2,
                                'Rarement'            :2,
                                'De temps en temps'   :3,
                                'Souvent'             :4,
                                ' Souvent'            :4,
                                'Tout le temps'       :5
                                },
                     new_var = 'bn70'
                     )
     #Relation avec les autres
    score_aggregate( row=row,
                     list_var = relation_avec_les_autres ,

                     dic_mod = {'La plupart du temps'           :1,
                                "Moins d'une fois par semaine"  :2,
                                "Moins d'une  fois par semaine" :2,
                                "Plus d'une fois par semaine"   :3,
                                'Presque jamais'                :4
                                },

                     new_var = 'bn2345'
                     )

     #relation avec la famille

    score_aggregate( row=row,
                     list_var =relation_avec_la_famille ,
                     dic_mod = {'Excellentes'     :1,
                                'Bonnes'          :2,
                                'Moyennes'        :3,
                                 'Mauvaises'      :4,
                                'Je ne sais pas'  :5,
                                },
                     new_var = 'cas3_0'
                     )


    return row

# Appliquer la fonction sur chaque ligne du DataFrame
data = data.apply(remplacer, axis=1)


# df, meta = pyreadstat.read_dta(file_path,encoding='latin1')

#Score de bien-être mental

for col in data.columns:
  type_ = data[col].dtype
  data[col] = data[col].astype(type_)


for col in New_loisir_variables+New_travail_variables:
  data[col] =  24*data[col]/data[New_loisir_variables+New_travail_variables].sum(axis=1, skipna=True)
```

```python
# Stocker les ensembles de données synthétiques dans un dictionnaire

variables_quantitatives = sensitive_variables_quanti + quasi_identifiant_quanti
sensitive_variables = sensitive_variables_quali + sensitive_variables_quanti
def evaluate_models(synthetic_datasets, real_data):
    import pandas as pd

    # Initialiser une liste pour stocker les résultats
    results_list = []

    # Fonction pour calculer les métriques pour un modèle donné
    def compute_metrics(real_data, synthetic_data, model_name):
        # Convert relevant columns to numeric type
        for col in real_data.columns:
            if col in variables_quantitatives:
                try:
                    real_data[col] = pd.to_numeric(real_data[col])
                    synthetic_data[col] = pd.to_numeric(synthetic_data[col])
                except:
                    pass

        # Ensure both real and synthetic data have the same column types
        for col in real_data.columns:
            if real_data[col].dtype != synthetic_data[col].dtype:
                try:
                    synthetic_data[col] = synthetic_data[col].astype(real_data[col].dtype)
                except:
                    pass

        metrics = {
            'Model': model_name,
            'CategoricalCAP': CategoricalCAP.compute(
                real_data#.replace([np.inf, -np.inf], "NA_CAP"),
                synthetic_data#.replace([np.inf, -np.inf], "NA_CAP"),
                key_fields=quasi_identifiants_quali,
                sensitive_fields=sensitive_variables
            ),
            'CSTest': CSTest.compute(real_data, synthetic_data),
            'KSComplement': KSComplement.compute(real_data, synthetic_data),
            'ContinuousKLDivergence': ContinuousKLDivergence.compute(real_data, synthetic_data),
            'DiscreteKLDivergence': DiscreteKLDivergence.compute(real_data, synthetic_data),
        }
        return metrics

    # Calculer les métriques pour chaque modèle
    for model_name, synthetic_data in synthetic_datasets.items():
      if model_name == 'SDC classique':
        synthetic_data = KanoLdiv_data[real_data.columns]
        # Calculer les métriques pour ce modèle
      metrics = compute_metrics(real_data, synthetic_data, model_name)
      results_list.append(metrics)

    # Convertir les résultats en DataFrame
    results_df = pd.DataFrame(results_list)

    return results_df
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.preprocessing import StandardScaler

def significance_stars(p_value):
    if p_value < 0.01:
        return '***'
    elif p_value < 0.05:
        return '**'
    elif p_value < 0.1:
        return '*'
    else:
        return ''

def regression_test(data, column_names, target_column='bn70', model_type='OLS'):
    # Handle missing or infinite values in the data
    d = data[column_names]
    d.replace([np.inf, -np.inf], np.nan, inplace=True)  # Replace infinite values with NaN
    d.dropna(inplace=True)  # Drop rows with missing values

    # Identify non-numeric columns in your DataFrame
    non_numeric_cols = d.select_dtypes(exclude=[np.number]).columns

    # Use one-hot encoding for non-numeric columns
    d = pd.get_dummies(d, columns=non_numeric_cols, drop_first=True).astype(float)

    # Scale numeric columns
    scaler = StandardScaler()
    d[d.columns] = scaler.fit_transform(d)

    # Split the data into features (X) and target variable (y)
    y = d[target_column]
    X = d.drop(target_column, axis=1)  # Select all columns except the target variable
    X = sm.add_constant(X)  # Add a constant (intercept)

    # Select and fit the model
    if model_type == 'OLS':
        model = sm.OLS(y, X).fit()
        adj_r_squared = model.rsquared_adj
    elif model_type == 'RLM':
        model = sm.RLM(y, X).fit()
        adj_r_squared = None  # RLM does not provide rsquared_adj
    elif model_type == 'GLM':
        model = sm.GLM(y, X).fit()
        adj_r_squared = None  # GLM does not provide rsquared_adj
    elif model_type == 'QuantReg':
        model = smf.quantreg(f'{target_column} ~ {" + ".join(column_names[:-1])}', data).fit()
        adj_r_squared = None  # QuantReg does not provide rsquared_adj

    # Extract coefficients and p-values
    coef_df = pd.DataFrame({
        'Coefficient': model.params,
        'P-Value': model.pvalues
    })

    # Add significance stars and round coefficients
    coef_df['Significance'] = coef_df['P-Value'].apply(significance_stars)
    coef_df['Coefficient'] = coef_df.apply(lambda x: f"{x['Coefficient']:.3f}{x['Significance']}", axis=1)

    return coef_df['Coefficient'], adj_r_squared

def compare_datasets(datasets, column_names, target_column='bn70', models=['OLS', 'RLM', 'GLM', 'QuantReg']):
    results = {}
    # Removed the separate index list as it was causing issues

    for model_type in models:
        model_results = {}
        for name, data in datasets.items():
            coef_df, adj_r_squared = regression_test(data, column_names, target_column, model_type)
            # Include Adj R² only for OLS
            if model_type == 'OLS':
                model_results[name] = [f"Adj R²: {adj_r_squared:.3f}"] + list(coef_df)
            else:
                model_results[name] = list(coef_df) # Directly use coef_df for other models
        # Use coef_df.index directly for creating the DataFrame
        print(name)
        results[model_type] = pd.DataFrame(model_results, index=coef_df.index)

    return results
```
